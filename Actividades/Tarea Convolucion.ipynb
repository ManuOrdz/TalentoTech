```python
import numpy as np 
import matplotlib.pyplot as plt
```

T = np.arange(0,100,0.1)
x = np.sin(60*T)+ 0.7*np.sin(700*T) + 0.1*np.random.normal(T)
h = np.ones(10)/10 #Mean 
h1 = np.array([1,2,3,2,1]) #Gaussian
h2 = np.array([1,-1]) # derivada
y = np.convolve(x,h,'same')
y1 = np.convolve(x,h1,'same')
y2 = np.convolve(x,h2,'same')

```python
_, ax = plt.subplots(ncols=2)
ax[0].plot(T[:100],x[:100])
ax[1].plot(T[:100],y[:100])
```

```python
_, ax = plt.subplots(ncols=2)
ax[0].plot(T[:100],x[:100])
ax[1].plot(T[:100],y1[:100])
```
